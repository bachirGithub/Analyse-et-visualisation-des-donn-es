```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
visuel=pd.read_csv("clean_data.csv");
```

## Distribution du status du prêt

> Dans ce graphique nous mettons en évidence la distribution des status des prêts, en corresondant chaque prêt à la fréquence qui lui ai propre.
> **Observations**: le prêt en cours présente la fréquence la plus élèvé pendant que les prêts en souffrance présentent une faible fréquence. Les prêt remboursés sont aussi bien représentés.

```python
def diagramme_barre(dataframe,col):
    couleur=sb.color_palette()[0]
    return sb.countplot(data=dataframe,x=col,color=couleur);
diagramme_barre(visuel,'LoanStatus')
plt.xticks(rotation=90);
plt.title('Distribution du status du prêt')
plt.xlabel('Status du prêt')
plt.ylabel('Fréquence');
```

## Relation status du prêt et status d'employé

> Ce graphique met en relation les status du prêt et les status d'employé et fait ressortir les différentes variations.
> **Observations**: Les employés detiennent les grande partie des prêts courants. Nous remaquons aussi que les travailleurs à temps plein ont une grande fréquence de remboursement.

```python
plt.figure(figsize=[14,6])
sb.countplot(data = visuel, x = 'LoanStatus', hue = 'EmploymentStatus', palette = 'Blues_r')
plt.xticks(rotation=90)
plt.xlabel("Satus du prêt")
plt.ylabel("Fréquence")
plt.title("Status du prêt selon les status des employés");
```

## Relation entre taux annuel effectif global, etat du prêt et status employé

>ce graphique montre la relation entr le taux mentuel effetif global avec l'etat du prêt et status employé. **Observations** : Les employés et les travailleurs à tant plein sont en général ceux qui ont le taux annuel effectif global le plus faible.
Pour le status de prêt completed les non salariés et les indépendants ont  le taux annuel effectif global le plus élevé.

```python
plt.figure(figsize=[15,5])
sb.boxplot(data=visuel,x="LoanStatus",y="BorrowerAPR",hue="EmploymentStatus");
plt.xticks(rotation =90);
plt.legend(ncol=4);

plt.title("Relation entre taux annuel effectif global, etat du prêt et status employé")
plt.ylabel("Taux annuel effectif global")
plt.xlabel("Etat du prêt");
```
