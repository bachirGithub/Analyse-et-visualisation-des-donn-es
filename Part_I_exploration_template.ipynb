## Exploration  Univariée

Je vais ici explorer la distribution des variables les unes après les autres, de la gauche vers la droite.

```python
def histogramme(dataframe,col,taille_bins):
    bins=np.arange(dataframe[col].min(),dataframe[col].max()+taille_bins,taille_bins)
    return plt.hist(data=dataframe,x=col,bins=bins);
histogramme(clean_loan,'Term',5);
#plt.hist(data=clean_loan,x='Term',bins=20);
```

>> **L'histogramme ici ne donne pas assez d'information pour tirer de conclusions assez précises.**

```python
# definisons une fonction pour le traçage de countplot
def diagramme_barre(dataframe,col):
    couleur=sb.color_palette()[0]
    return sb.countplot(data=dataframe,x=col,color=couleur);
diagramme_barre(clean_loan,'Term')
plt.title('La durée du prêt par mois')
plt.xlabel('Durée en mois')
plt.ylabel('Fréquence');
#sb.countplot(data=clean_loan,x='Term',color=sb.color_palette()[0]);
```

>> **Nous remarquons que la durée du prêt est de trois types à savoir 36,60 et 12**

```python
diagramme_barre(clean_loan,'LoanStatus')
plt.xticks(rotation=90);
plt.title('Distribution du status du prêt')
plt.xlabel('Status du prêt')
plt.ylabel('Fréquence');
```

>> **Le prêt en cours présente la fréquence la plus élèvé pendant que les prêts en souffrance présentent une faible fréquence. Les prêt remboursés sont aussi bien représentés.**

```python
histogramme(clean_loan,'BorrowerAPR',0.008)
plt.title('distribution du taux annuel effectif global')
plt.xlabel('Taux')
plt.ylabel('Fréquence');
```

>> **Le diagramme présente une variation multimodal avec des picks au niveau de 0.9,0.20,0.3 et 0.35. Représentant, selon le dictionnaire des données, le taux annuel effectif global (APR) de l'emprunteur pour le prêt. Nous remarquons que la fréquence la plus élevée se trouve au tour de 0.35.**

```python
clean_loan.BorrowerAPR.value_counts()
```

>> **La valeur la plus fréquente est précisement 0.35797 suivie de 0.35643.**

```python
histogramme(clean_loan,'BorrowerRate',0.0089);
plt.title("Distribution du taux d'emprunteur")
plt.xlabel("Taux emprunteur")
plt.ylabel("Fréquence");
```

>> **BorrowerRate représente, selon le dictionnaire des données, le taux d'intérêt de l'emprunteur pour ce prêt.
Son diagramme montre une variation bimodal avec des pick dans la zone de 0.15 et de 0.32 qui est la fréquence plus élevée.**

```python
clean_loan.BorrowerRate.value_counts()
```

>> **La fréquence la plus élevée est précisement 0.3177 suivie de 0.3199**

```python
plt.hist(data=clean_loan, x='ProsperRating (Numeric)');
```

>> **La note moyenne C est la plus fréquente dans le dataset suivie de la note B. La AA (plus forte) est la moins fréquente.**

## Bivariate Exploration

> Commençons par voir les rélations entre variables.

```python
numeric_vars = ['Term', 'BorrowerRate', 'DelinquenciesLast7Years', 'StatedMonthlyIncome','ProsperRating (numeric)','LoanOriginalAmount', 'Recommendations', 'Investors']
categoric_vars = ['LoanStatus', 'BorrowerAPR','ProsperRating (Alpha)','ListingCategory (numeric)','EmploymentStatus', 'Annee','mois']
```

Pour les variables numeriques.

```python
# Matrix de correlation
plt.figure(figsize = [8, 5])
sb.heatmap(clean_loan[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
plt.show()
```

La plus part des correlation entre variables sont faibles.

On remarque qu'en même quelque correlations moyennes entre **LoanoriginalAmount** et des variables comme **ProsperRating (numeric)**, **Term**, et **Investors**, aussi entre **Investors** et **ProsperRating (numeric)**.

On observe aussi une forte correlation negative entre **ProsperRating (numeric)** et **BorrowerRate**.

```python
g = sb.PairGrid(data = clean_loan.sample(500), vars = numeric_vars)
g = g.map_diag(plt.hist, bins = 20)
g.map_offdiag(plt.scatter);
```

```python
sb.regplot(data=clean_loan,y='Investors',x="ProsperRating (numeric)")
plt.xlabel("Nombre d'investisseurs")
plt.ylabel("Taux de prosperité")
plt.title("Distribution des investisseur selon la taux de prospérité");
```

Nous observons une forme d'augmentation du taux de prospérité en fonction du nombre d'investisseurs.

```python
plt.figure(figsize=[14,6])
sb.countplot(data = clean_loan, x = 'LoanStatus', hue = 'EmploymentStatus', palette = 'Blues_r')
plt.xticks(rotation=90)
plt.xlabel("Satus du prêt")
plt.ylabel("Fréquence")
plt.title("Status du prêt selon les status des employés");
```

Les employés detiennent les grande partie des prêts courants. Nous remaquons aussi que les travailleurs à temps plein ont une grande fréquence de remboursement.

```python

plt.scatter(data = clean_loan, x = 'ProsperRating (numeric)', y = 'LoanOriginalAmount',cmap='viridis_r')
plt.xlabel('Montant original du prêt')
plt.ylabel('Taux de prospérité')
plt.title("Relation du taux de propérité avec le montant original du prêt");
```

Nous remarquons que le taux de prosperité a tendance à croître avec la croissance du montant original du prêt.

```python
sb.boxplot(data=clean_loan,x="Term",y="BorrowerAPR",color=sb.color_palette()[0])

plt.xlabel("Durée du prêt en mois'");
plt.ylabel("Taux annuel effectif global");
plt.title("Relation entre la durée du prêt et le taux annuel effectif global");
```

Le prêt à durée de 36 mois à une fréquence beaucoup plus large.

## Multivariate Exploration

```python
sb.catplot(x='ProsperRating (Alpha)', hue='LoanStatus', col='Term',data=clean_loan, kind='count', palette='Blues_r', col_wrap=1,height=5, aspect=2)
plt.title('Rapport taux de prospérité et Status du prêt par durée du prêt');
```

Pour la durée de 36 mois le taux de prospérité de type A semble être le mieux représenté avec une domination du status de prêt Current suivie de compled.

Pour la durée de 60 mois le taux de prospérité de type C présente le pick le plus élevé pour le status de prêt de type Current et completed.

Pour la durée de 12, qui est la moins représente, on constate plutôt une repartition équitable entre les fréquente des prêts type Current pour les différents types de taux.

```python
plt.figure(figsize=[15,5])
sb.boxplot(data=clean_loan,x="LoanStatus",y="BorrowerAPR",hue="EmploymentStatus");
plt.xticks(rotation =30);
plt.legend(ncol=4);

plt.title("Relation entre taux annuel effectif global, etat du prêt et status employé")
plt.ylabel("Taux annuel effectif global")
plt.xlabel("Etat du prêt");
```

Les employés et les travailleurs à tant plein sont en général ceux qui ont le taux annuel effectif global le plus faible.
Pour le status de prêt completed les non salariés et les indépendants ont  le taux annuel effectif global le plus élevé.
